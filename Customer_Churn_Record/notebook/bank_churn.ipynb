### Analize

```python
ditribusi_churn = churn_df['Exited'].value_counts(normalize=True) * 100
print(ditribusi_churn)
```

# Visualiasi

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Exited', data=churn_df)
plt.title('Distribusi Nasabah Churn vs Tidak Churn')
plt.xlabel('Status Churn')
plt.ylabel('Jumlah Nasabah')
```

#### Jumlah Pasti

```python
total_rows = len(churn_df)

jumlah_churn = (ditribusi_churn / 100) * total_rows

jumlah_churn = jumlah_churn.round(0).astype(int)
print(jumlah_churn)
```
